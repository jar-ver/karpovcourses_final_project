<p>Напишите код на python для получения нужной таблицы и укажите 5 наиболее распространённых паттернов.</p>

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
```

```python
df = pd.read_csv('https://stepik.org/media/attachments/lesson/409319/test1_completed.csv')
```

```python
df.head()
```

```python
df_pivot = pd.pivot(df, values='Количество', index='id', columns='Товар')\
             .fillna(0)\
             .applymap(lambda x: False if x == 0 else True)
```

```python
df_pivot.head()
```

```python
# http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/

frequent_itemsets = apriori(df_pivot, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.head()
```

```python
# нам нужны пары
frequent_itemsets = frequent_itemsets.query('length == 2')
```

```python
frequent_itemsets['itemsets'] = frequent_itemsets.itemsets.apply(lambda x: list(x))
frequent_itemsets.sort_values(by='support', ascending=False).head()
```

<img src='https://www.analyticsindiamag.com/wp-content/uploads/2019/06/support.png' align="left">

```python
# "Total number of transactions"
transact_number = df.id.nunique()
```

```python
# "support" -> "встречаемость"
frequent_itemsets["Встречаемость"] = (frequent_itemsets.support * transact_number).astype(int)
```

```python
frequent_itemsets["1_Товар"] = frequent_itemsets.itemsets.map(lambda x: x[0])
frequent_itemsets["2_Товар"] = frequent_itemsets.itemsets.map(lambda x: x[1])
```

```python
top_5 = frequent_itemsets[['1_Товар', '2_Товар', 'Встречаемость']]\
                            .sort_values('Встречаемость', ascending=False)\
                            .head()
```

```python
top_5
```

```python
top_5.to_csv('task_1_top_5.csv', index=False)
```

<p>Получили достатоно логичные и очевидные комбинации товаров, возможно, стоит взглянуть на менее распространённые паттерны.</p>

База данных доступна по имени project_variant_2</p>

```python
import pandahouse as ph
import pandas as pd
```

```python
query = """
SELECT
  purchase.purchase_date AS purchase_date,
  purchase.purchase_id AS purchase_id,
  purchase.client_id AS client_id,
  dateDiff('year', client.birth_date, toDate(now())) AS client_age,
  dateDiff('year', client.registration, toDate(now())) AS client_registration_age,
  categories.client_category AS client_category,
  promotion.promotion_name AS promotion_name,
  promotion.category_name AS category_name,
  promotion.partner_name AS partner_name,
  city_clients.client_city AS client_city,
  city_purchases.city AS city,
  toFloat32(purchase.price * purchase.quantity) AS revenue,
  toInt8(purchase.quantity) AS quantity
FROM
  project_variant_2.purchase
  LEFT JOIN project_variant_2.client ON client.client_id = purchase.client_id
  LEFT JOIN project_variant_2.promotion ON purchase.promotion_id = promotion.promotion_id
  LEFT JOIN (
    SELECT
      DISTINCT client_city_id,
      client_city
    FROM
      project_variant_2.city
  ) AS city_clients ON client.client_city_id = city_clients.client_city_id
  LEFT JOIN (
    SELECT
      DISTINCT city_id,
      city
    FROM
      project_variant_2.city
  ) AS city_purchases ON purchase.city_id = city_purchases.city_id
  LEFT JOIN (
    SELECT
      client_id,
      if(purchases = 1, 'new', 'old') AS client_category
    FROM
      (
        SELECT
          client_id,
          uniqExact(purchase_id) AS purchases
        FROM
          project_variant_2.purchase
        GROUP BY
          client_id
      )
  ) AS categories ON purchase.client_id = categories.client_id
WHERE
  purchase.status = '1'
  AND purchase.purchase_date BETWEEN '2020-05-01'
  AND '2020-08-01'
"""
```
